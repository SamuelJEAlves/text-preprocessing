```python
import os
import pandas as pd
from tika import parser

pasta_pdf = "docs"

dados = []

for arquivo_pdf in os.listdir(pasta_pdf):
    if arquivo_pdf.endswith('.pdf'):
        caminho = os.path.join(pasta_pdf, arquivo_pdf)
        
        parsed_pdf = parser.from_file(caminho)
        
        texto = parsed_pdf.get('content', '')

        dados.append({
            "texto": texto,
        })

df = pd.DataFrame(dados)
display(df)
```

```python
import nltk
from nltk.corpus import stopwords
from unidecode import unidecode
import re

nltk.download('punkt')
nltk.download('stopwords')

stopwords_pt = set(stopwords.words('portuguese'))
stopwords_en = set(stopwords.words('english'))
stopwords_es = set(stopwords.words('spanish'))

custom_stopwords = {'nao', 'sim', 'wikipedia', 'org', ',', "'", 'enciclopedia', 'oul', 'atl', 'tico', 'wiki', 'png', 'kit'}

stopwords_full = stopwords_pt.union(stopwords_en).union(stopwords_es).union(custom_stopwords)
```

```python
def preprocess_text(text):
    text = unidecode(text)  # Remove acentos
    text = re.sub(r'[^\w\s]', ' ', text) # Remove pontuação
    text = re.sub(r'\s+', ' ', text).strip() # Remove espaços extras
    text = re.sub(r'\W+', ' ', text)  # Remove pontuação e caracteres especiais
    text = re.sub(r'\d+', ' ', text)  # Remove números
    text = re.sub(r'\n+', ' ', text) # Remove quebras de linha
    text = re.sub(r'_+', ' ', text) # Remove underline
    text = re.sub(r'(http|https|www|html|index|htm)', '', text)  # Remove URL e termos comuns de sites
    tokens = nltk.word_tokenize(text.lower())  # Tokenização e normalização de caixa
    tokens = [word for word in tokens if word not in stopwords_full and len(word) > 2 and not re.match(r'^[a-zA-Z]$', word)]  # Remove stopwords e palavras curtas
    return tokens
```

```python
df['texto'] = df['texto'].apply(preprocess_text)
df
```

```python
df['texto'] = df['texto'].astype(str).fillna('')
```
